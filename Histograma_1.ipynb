#Histograma

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats #funciones de estadistica
```

```python
datos = np.loadtxt('tiempos.txt')
cant_datos = 600
```

```python
media = datos.mean()
desviacion_estandar = datos.std()
print('Media: ', media)
```

#### es para ver la aletoriedad de los datos o si hay alguna tendencia

```python
plt.figure(figsize=(14,6))
plt.plot(datos, 'o-', ms=4, lw=1) #plt.plot(datos"es lo que ongresas", 'o-' "es la linea", ms=4 "tamaño de puntos", lw=1 "tamaño de la linea")
plt.xlabel('Número de medición', fontsize=14)
plt.ylabel('T [s]', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,6)) #plt.plot(datos, 'p')
plt.plot(datos, 'o', ms=4, lw=1) #para que no tenga lineas
plt.xlabel('titulo de abajo')
plt.ylabel('titulo lateral')
plt.grid()
plt.title('titulo que se repite 2 veces')
```

```python
import random
random.shuffle(datos) #mezclo los datos
plt.figure(figsize=(14,6)) #plt.plot(datos, 'p')
plt.plot(datos, 'o', ms=4, lw=1) #para que no tenga lineas
plt.xlabel('titulo de abajo')
plt.ylabel('titulo lateral')
plt.grid()
plt.title('titulo que se repite 2 veces')
```

```python
media = datos.mean()
desv = datos.std()
rango = media-desv, media+desv

# separo los datos en dos grupos: dentro y fuera del rango que quiero
lugares_bajo = datos<rango[0]
lugares_sobre = datos>rango[1]
lugares_fuera = np.logical_or(lugares_bajo, lugares_sobre)
lugares_dentro = np.logical_not(lugares_fuera)

# ¿Qué porcentaje de los datos cae dentro del rango?
# (¿Qué porcentaje esperamos, dado que la distribuación es gaussiana?)
porcent = f'{np.sum(lugares_dentro/cant_datos)*100:.1f}% de los datos'

# grafico datos de nuevo

plt.plot(np.arange(cant_datos)[lugares_dentro], datos[lugares_dentro], 'o', color='C0', label=porcent, ms=4)
plt.plot(np.arange(cant_datos)[lugares_fuera], datos[lugares_fuera], 'o', color='0.6', ms = 4)
plt.xlabel('# de medición')
plt.ylabel('valor[unidades]')
plt.grid()
plt.title('Valores de todos los datos del experimento')
plt.legend()

# agrego la línea hjorizontal de la media
x_range = plt.xlim() #me guardo los límites actuales dle gráfico
plt.hlines(datos.mean(), xmin=x_range[0], xmax=x_range[1], ls="--", label = "Media")

# y la barra con el intervalo [μ-σ, μ+σ]:
plt.fill_between(x_range, *rango, alpha=0.5)
plt.xlim(x_range)
```

```python
import math as ma
import numpy as np

k = ma.log(cant_datos,10)
print(k)
```

```python
import math
from math import log
logaritmo_datos = log(cant_datos,10) #log(numero,base)
cant_intervalos = 1 + 3.322*logaritmo_datos
print('cantidad_de_intervalos', cant_intervalos)
print('Máximo:',max(datos))
print('Mímino:',min(datos))
x_max = max(datos)
x_min = min(datos)
ancho = (x_max - x_min)/k
print('ancho_columna: ', ancho)
```
